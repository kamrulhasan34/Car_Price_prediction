```python
def get_brand_name(car_name):
    if isinstance(car_name, str):  # Check if car_name is a string
        car_name = car_name.split(' ')[0]
        return car_name.strip()
    else:
        return car_name
```

```python
def clean_data(value):
    if isinstance(value, str):  # Check if value is a string
        value = value.split(' ')[0]
        value = value.strip()
        if value == '':
            return 0.0  # Return 0.0 for empty strings
        else:
            return float(value)
    else:
        return value
```

```python
input_data=cars.drop(columns=['selling_price'])
output_data=cars['selling_price']
```

```python
x_train, x_test, y_train, y_test= train_test_split(input_data, output_data, test_size=0.2)
```

```python
x_test = x_test.map(clean_data)
```

```python
#model creation
model= LinearRegression()
```

```python
import re

def clean_data(value):
    if isinstance(value, str):
        value = re.sub(r'[^\d\.]+', '', value)  # Remove non-numeric characters
        value = value.strip()
        if value == '':
            value = 0.0
        else:
            value = float(value)
    return value

x_train = x_train.map(clean_data)
x_test = x_test.map(clean_data)

model.fit(x_train, y_train)
predict = model.predict(x_test)
```
